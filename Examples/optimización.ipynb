## Definimos 5 funciones analiticas con optimos globales

```python
# Se define las variables simbólicas
x, y = sympy.symbols('x y')

# 1. Función cuadrática simple (convexa)
f1 = x**2 + 2*y**2

# 2. Función de Rosenbrock (no convexa, con un mínimo global)
f2 = (1 - x)**2 + 100 * (y - x**2)**2

# 3. Función de Himmelblau (no convexa, con cuatro mínimos locales)
f3 = (x**2 + y - 11)**2 + (x + y**2 - 7)**2

# 4. Función de Beale (no convexa, con un mínimo global)
f4 = (1.5 - x + x*y)**2 + (2.25 - x + x*y**2)**2 + (2.625 - x + x*y**3)**2

# 5. Función de Booth (convexa, con un mínimo global)
f5 = (x + 2*y - 7)**2 + (2*x + y - 5)**2
```

# **Ejemplo 1 : Optimización del consumo de gasolina de un vehiculo**

La velocidad de un vehículo juega un papel crucial en la optimización del consumo de gasolina. El consumo de combustible generalmente no es una relación lineal con la velocidad, sino que tiene una forma parabólica, donde el consumo es bajo a velocidades muy bajas, alcanza un mínimo en una velocidad media y luego aumenta nuevamente a altas velocidades. Esto se debe a la resistencia al aire (que aumenta con la velocidad) y la eficiencia del motor (que puede disminuir a altas revoluciones).

Imagina que estás conduciendo un vehículo y deseas minimizar el consumo de combustible para un viaje específico. Sabemos que el consumo de combustible por unidad de distancia no es constante y depende de la velocidad 𝑣, según la siguiente relación:

$$ C(v) = a v^2 + b v + c $$

donde:

* 𝐶(𝑣) es el consumo de combustible en litros por kilómetro.
* 𝑣 es la velocidad del vehículo en km/h.
* 𝑎, 𝑏 y 𝑐 son parámetros que dependen de características del vehículo, como la eficiencia del motor, la resistencia al viento, la fricción de los neumáticos, etc.

El objetivo es encontrar la velocidad 𝑣 que minimiza 𝐶(𝑣).

# **Como abordamos el problema de optimización de forma analítica?**

Para minimizar el consumo de combustible 𝐶(𝑣), necesitamos encontrar el valor de 𝑣 que minimiza esta función cuadrática. La derivada de 𝐶(𝑣) con respecto a 𝑣 nos da la tasa de cambio del consumo de combustible respecto a la velocidad. Para encontrar el valor de 𝑣 que minimiza 𝐶(𝑣), igualamos la derivada a cero.

La derivada de 𝐶(𝑣) con respecto a 𝑣 es:

$$ \frac{d}{dv} C(v) = 2a v + b $$

Igualamos a cero para encontrar el valor de 𝑣:

$$ 2a v + b = 0 $$

Despejamos 𝑣 :

$$ v = \frac{-b}{2a} $$

Este es el valor de la velocidad 𝑣 que minimiza el consumo de combustible.

# **Que pasa si tenemos funciones más complejas donde la solución análitica no es posible?**

```python
# Funcion Ackley

f6 = -2 * sympy.exp(-1 * sympy.sqrt(0.5 * (x**2 + y**2))) - \
     sympy.exp(0.5 * (sympy.cos(0.5 * x) + sympy.cos(0.5 * y))) + 2 + sympy.exp(1)

plot3d(f6, (x, -5, 5), (y, -5, 5), title='Función de Ackley', xlabel='x', ylabel='y', zlabel='f(x,y)')
```

```python
# Función Rastrigin
f7 = 20 + x**2 - 10 * sympy.cos(2 * sympy.pi * x) + y**2 - 10 * sympy.cos(2 * sympy.pi * y)
plot3d(f7, (x, -5.12, 5.12), (y, -5.12, 5.12), title='Función Rastrigin', xlabel='x', ylabel='y', zlabel='f(x,y)')
```

```python
# Gaussian con Oscilaciones

a = 3       # Amplitude
b = 0.5     # Desviación estandar
c = 0       # x- del centro
d = 0       # y- del centro
omega = 1   # Frecuencia de oscilación

f8 = a * sympy.exp(-((x - c)**2 + (y - d)**2) / (2 * b**2)) + sympy.sin(omega * x) * sympy.cos(omega * y)

plot3d(f8, (x, -5, 5), (y, -5, 5), title='Gaussian con Oscilaciones', xlabel='x', ylabel='y', zlabel='f(x,y)')
```

# **Opción más intuitiva: Fuerza bruta**



```python
%%time
import numpy as np

# Optimización por fuerza bruta
max_f8 = float('-inf')

step = 0.3  # Tamaño del paso de exploración
x_range = np.arange(-60, 60, step) # Rango del espacio de busqueda X
y_range = np.arange(-60, 60, step) # Rango del espacio de busqueda Y
print(x_range.shape)
print(y_range.shape)
# Se recorre todo el espacio de busqueda a fuerza bruta
for i in x_range:
  for j in y_range:
    x_val = i
    y_val = j
    f8_val = f8.subs({x: x_val, y: y_val})
    if f8_val > max_f8:
      max_f8 = f8_val
      optimal_x = x_val
      optimal_y = y_val

print(f"Valor máximo de f8: {max_f8}")
print(f"x optimo: {optimal_x}")
print(f"y optimo: {optimal_y}")
```

# **Otra opción: Optimización de Monte Carlo**

```python
%%time
import sympy
import numpy as np
import random

# Monte Carlo
def monte_carlo_max(func, x_range, y_range, num_samples):
    max_f8 = float('-inf')
    optimal_x = None
    optimal_y = None

    for _ in range(num_samples):
        x_val = random.uniform(x_range[0], x_range[1])
        y_val = random.uniform(y_range[0], y_range[1])
        f8_val = func.subs({x: x_val, y: y_val})

        if f8_val > max_f8:
            max_f8 = f8_val
            optimal_x = x_val
            optimal_y = y_val

    return max_f8, optimal_x, optimal_y

# Definimos el espacio de busqueda
x_range = (-10, 10)
y_range = (-10, 10)
num_samples = 100000

# Optimización de Monte Carlo
max_f8, optimal_x, optimal_y = monte_carlo_max(f8, x_range, y_range, num_samples)

print(f"Valor máximo de f8 (Monte Carlo): {max_f8}")
print(f"x óptimo (Monte Carlo): {optimal_x}")
print(f"y óptimo (Monte Carlo): {optimal_y}")
```

# **Una mejor opción: Algoritmo de gradiente ascendente**

```python
%%time
# Algoritmo de Gradiente Ascendente
def gradient_ascendente(func, initial_x, initial_y, learning_rate=0.01, iterations=1000):
    x_current, y_current = initial_x, initial_y

    # Se calculan los gradiente en la función simbolica
    gradient_x = sympy.diff(f8, x)
    gradient_y = sympy.diff(f8, y)

    # Se convierten los gradientes simbolicos en númericos
    gradient_x_lambda = sympy.lambdify((x, y), gradient_x, 'numpy')
    gradient_y_lambda = sympy.lambdify((x, y), gradient_y, 'numpy')

    for _ in range(iterations):
        grad_x = gradient_x_lambda(x_current, y_current)
        grad_y = gradient_y_lambda(x_current, y_current)
        x_current = x_current + learning_rate * grad_x
        y_current = y_current + learning_rate * grad_y

    return x_current, y_current, func.subs({x: x_current, y: y_current})

initial_x, initial_y = (2, 2)  # Punto inicial
learning_rate = 0.01  # Tasa de aprendizaje
iterations = 1000      # Número de iteraciones

optimal_x, optimal_y, optimal_f8 = gradient_ascendente(f8, initial_x, initial_y, learning_rate, iterations)

print(f"x optimo: {optimal_x}")
print(f"y optimo: {optimal_y}")
print(f"f8 máximo: {optimal_f8}")
```
